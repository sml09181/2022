```python
def gender_features2(name):
    features = {}
    features["first_letter"] = name[0].lower()
    features["last_letter"] = name[-1].lower()
    for letter in 'abcdefghijklmnopqrstuvwxyz':
        features["count({})".format(letter)] = name.lower().count(letter)
        features["has({})".format(letter)] = (letter in name.lower())
    return features
```

```python
import nltk
from nltk.corpus import brown
```

```python
suffix_fdist = nltk.FreqDist()
for word in brown.words():
    word = word.lower()
    suffix_fdist[word[-1:]] += 1
    suffix_fdist[word[-2:]] += 1
    suffix_fdist[word[-3:]] += 1
common_suffixes = [suffix for (suffix, count) in suffix_fdist.most_common(100)]
```

```python
suffix_fdist
```

```python
def pos_features(word):
    features = {}
    for suffix in common_suffixes:
        features['endswith({})'.format(suffix)] = word.lower().endswith(suffix)
    return features
```

```python
tagged_words = brown.tagged_words(categories='news')
featuresets = [(pos_features(n), g) for (n, g) in tagged_words]
size = int(len(featuresets) * 0.1)
```

```python
classifier = nltk.DecisionTreeClassifier.train(train_set)
print("POS Tagging:", nltk.classify.accuracy(classifier, test_set))
```

### Plus. Exploiting Context

```python
def pos_features(sentence, i): 
    features = {"suffix(1)": sentence[i][-1:],
                "suffix(2)": sentence[i][-2:],
                "suffix(3)": sentence[i][-3:]}
    if i == 0:
        features["prev-word"] = "<START>"
    else:
        features["prev-word"] = sentence[i-1]
    return features
```

```python
pos_features(brown.sents()[0], 8)
```

```python
tagged_sents = brown.tagged_sents(categories='news')
featuresets = []
for tagged_sent in tagged_sents:
     untagged_sent = nltk.tag.untag(tagged_sent)
     for i, (word, tag) in enumerate(tagged_sent):
         featuresets.append( (pos_features(untagged_sent, i), tag) )
```

```python
size = int(len(featuresets) * 0.1)
train_set, test_set = featuresets[size:], featuresets[:size]
train_set[:4]
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set)
nltk.classify.accuracy(classifier, test_set)
```

```python
def segment_sentences(words):
    start = 0
    sents = []
    for i, word in enumerate(words):
        if word in '.?!' and classifier.classify(punct_features(words, i)) == True:
            sents.append(words[start:i+1])
            start = i+1
    if start < len(words):
        sents.append(words[start:])
    return sents
```

```python
sents = nltk.corpus.treebank_raw.sents()
tokens = []
boundaries = set()
offset = 0
for sent in sents:
        tokens.extend(sent)
        offset += len(sent)
        boundaries.add(offset-1)
```

```python
def punct_features(tokens, i):
        return {'next-word-capitalized': tokens[i+1][0].isupper(),
                'prev-word': tokens[i-1].lower(),
                'punct': tokens[i],
                'prev-word-is-one-char': len(tokens[i-1]) == 1}
```

```python
featuresets = [(punct_features(tokens, i), (i in boundaries))
        for i in range(1, len(tokens)-1)
                if tokens[i] in '.?!']
size = int(len(featuresets) * 0.1)
train_set, test_set = featuresets[size:], featuresets[:size]
classifier = nltk.NaiveBayesClassifier.train(train_set)
nltk.classify.accuracy(classifier, test_set)
print("Sentence Segmentation:", nltk.classify.accuracy(classifier, test_set))
```

#### PLUS 
prev-word라는 feature은 예측에 별로 도움이 되지 않는 걸로 보인다. 사실인지 확인하기 위해 random_state의 값에 따른 전반적인 accuracy_score 그래프를 그려 보자.

```python
model_Y = BernoulliNB()
model_N = BernoulliNB()
Y = list()
N = []
for i in range(45):
    X_train_Y, X_test_Y, y_train_Y, y_test_Y = train_test_split(df.drop(labels='target', axis=1), df['target'], test_size = 0.1, random_state=i)
    X_train_N, X_test_N, y_train_N, y_test_N = train_test_split(df.drop(labels=['target', 'prev-word'], axis=1), df['target'], test_size = 0.1, random_state=i)
    model_Y.fit(X_train_Y, y_train_Y)
    model_N.fit(X_train_N, y_train_N)
    y_pred_Y = model_Y.predict(X_test_Y)
    y_pred_N = model_N.predict(X_test_N)
    Y.append(accuracy_score(y_test_Y, y_pred_Y))
    N.append(accuracy_score(y_test_N, y_pred_N))
check = pd.DataFrame({'Y': Y, 'N': N})
check.plot(figsize=(14,5), lw = 4.5, alpha = 0.8, color = ['cornflowerblue', 'lightcoral'])
```

거의 흡사하지만 prev-word feature를 고려하는 것이 몇몇 지점에서 더 높은 accuracy_score를 보인다.
